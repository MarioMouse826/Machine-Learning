# What is Machine Learning?

*"It is the field of study that gives computers the ability to learn without being explicitly programmed." *
   - Arthur Samuel, 1959

"A computer program is said to learn from Experience *E* with respect to some task *T* and some performance measure *P*, if its performance on *T*, as measured by *P*, improves with experience *E*."
   - Tom Mitchell, 1997

Machine Learning is great for:



*   Problems for which existing solutions require a lot of work and maintenance, such as long lists of rules (ML models can often simplify code and perform better than the traditional approach)

*   Complex problems for which using a traditional approach yields no good solution (the best machine learning techniques can perhaps find a solution)

*   Fluctuating environments (a machine learning system can easily be retrained on new data, always keeping it up to date)
*   Getting insights about complex problems and large amounts of data

*   The examples that the system uses to learn are called the **Training Set**
*   Each training example is called a **Training Instance or Sample**


*   The part of a Machine Learning system that learns and makes predictions is called a **Model**
*   **Data Mining** is what ML excels at because it involves digging into large amounts of data to discover hidden patterns

*   How are they guided during training? (Supervised, Unsupervised, Semi-Supervised, Self-Supervised, Reinforcement Learning, etc)

# Learning Paradigm



# Supervised Learning



*   The training set you feed to the learning algorithm includes the desired solutions (aka labels)
*  Classification (ex. Spam filters being trained with many example emails given a class already - hence the hypothesis must classify new emails per the learning algorithm) and
*  Regression (ex. Predicting a target
(already labelled) numeric value given aset of features - to train this sytem you also need many examples of items with their labelled features)



# Unsupervised Learning



*   Training data is unlabelled so the system learns without a teacher
*   Clustering to detect groups of similar items based on features is one example


*   Visualization algorithms are good examples of unsupervised learning: you feed a lot of complex and unlabelled data and they output a 2D or 3D representation of your data that can be easily plotted - Visualization algorithms typically try to preserve as much structure as they can so the viewer can understand how the data is organized and identify unsuspected patterns
*   Dimensionality Reduction is a related task to unsupervised learning (simplify data without losing necessary information). When doing feature extraction, it is often a good idea to try to reduce the number of dimensions in your training data using some dimensionality reduction algorithm before feeding it to your learning algorithm. Data will take up less disk and memory space, it will run much faster, and in many cases perform better.



*   Anomaly Detection is another example of an unsupervised task - system is trained on normal instances of for example, credit card detection, so it learns to recognize those, whiel new instances are recognized as either normal or anomalies. Similarly novelty detection aims to detect new instances that look different from previous instances in the training set
*   Association Rule Learning is another task where the goal is to dig into large amounts of data and discover interesting relations between attributes hence you might wish to put these items closer to one another.

# Self-Supervised Learning



*   Actually generating a fully labeled dataset from a fully unlabeled one. Once this dataset is labeled, only then can you use a supervised learning algorithm.
*   One example is image recognition or image reconstruction in computer vision - you can mask a small part of an image and retrain a model to reconstruct the missing part. During training, masked images are used as the inputs to the model and original images are used as the labels. (In simpler terms, you give the model an unlabeled dataset and you make the model reconstruct the data so that it generates the labels itself from scratch. - Both Computer Vision and NLP tasks tends to work this way.)


*   More often than not, models trained using self-supervised learning are not the final goal. You usually want to tweak and fine-tune the model for a slightly different task - one that you actually care about.
*  Transferring knowledge from one task to another is called transfer learning  - this is an important task when you are using deep neural networks. LLMs and other Transformer based methods are trained in a similar way to self-supervised learning (masking random words in a huge text corpus and training the model to predict the missing words - you can then fine-tune this pretrained model to various applications from sentiment analysis to chatbots)

This is much closer to supervised learning due to the fact that it uses labels (self-supervised learning generally focuses on the same tasks as supervised learning like classification and regression and not clustering, dimensionality reduction or anomaly detection).

# Training

# Batch Learning



*   The system must be trained using all available data. This will generally take a lot of time and computing resources, so it is typically done offline (hence offline learning)
*   System is trained and then launched into production without learning anymore; it just applies and generalizes based on what it has learned

* Model Rot aka data drift is a serious issues where a model's performance tends to decay over time because the world is dynamic and the model trained on Batch Learning is static by definition.

* You can solve this by regularly retraining the model on up-to-date data and how often you need to do that depends on the use case (financial markets are a fast-evolving system while classifying cats and dogs will not alter performance nearly as rapidly)

* If you want a batch learning system to know about new data (like a new type of spam, then you need to train a new version of the system form scratch on the full dataset including both new and old data), then replace the new model with the old one.

* This solution of training evaluating and launching a machine learning system can be automated so long as you update the data and train a new version as often as needed but this is extremely costly and will not adapt quickly to rapidly changing data. Training onthe full set of data requires extensive computing resources (CPU, memory space, disk space, disk I/O, etc). If you have a lot of data and need to automate the system to train from scratch every day, it will end up costing a lot of money. If the amount of data is huge and the system must always be up to date, batch learning is impossible.



# Online Learning



*   Train the system incrementally by feeding it data instances sequentially, either individually or in small groups called mini-batches. each learning step is fast and cheap so the system can learn about new data on the fly, as it arrives. The most common online algorithm by far is gradient descent.
*  Extremely useful for systems that need to adapt to change extremely rapidly ie financial patterns in the stock market. Also a good option for those with limited computing resources ie mobile devices. Most importantly, online algorithms can be used to train on huge datasets that cannot fit in one machine's memory (out of core learning). Algorithm loads part of the data, runs a training step on that data and repeats the process until it has run on all of the data.
* Note that higher learning rates are not always better - in online learning systems the system can rapidly adapt to new data but it will also tend to quickly forget the old data which ends with catastrophic forgetting (catastrophic interference). You don't want a spam filter to flag on only the latest kinds of spam that it was shown for instance.
Conversely, a learning rate that is too low will have more inertia (system will learn too slowly) however it will be less sensitive to noise or sequences in nonrepresentative outliers.

* Think of Online Learning more like "Incremental Learning"



# Modeling

# Instance-Based



*   System learns examples by heart as they are fed into the model and then it generalizes to new cases by using a similarity measure to compare them to the learned examples or a subset of them.
*   This is very trivial but shines best with smaller datasets, especially if data keeps changing but this does not scale very well at all. It requires deploying a whole copy of the training set to production; making predictions requires searching for similar instances, which can be quite slow; and it does not work with high-dimensional data such as images.



# Model-Based



*   You can generalize from a set of exapmles to build a model from these examples and use that model to make predictions - this is called model-based learning
*  How to know which values will make the model perform best? Specify a performance measure (utility function to specify how good the model is, or loss function to measure how bad it is).
* For Linear Regression you can feed training examples to the learning algorithm and the hypothesis finds the parameters that make the linear model fit best to the training data (this is what training the model means)



*   Other methods include ensemble learning, federated learnig, meta learning, etc.




# Main Challenges in Machine Learning



# Insufficient Quality of Training Data

 - ML requires a LOT of data for most algorithms to work properly. Even for very simple problems you typically need thousands of examples, and for complex problems such as image or speech recognition you may need millions of examples. Ex. Banko and Brill, 2001 showed that different ML algorithms even the simple ones perform almost identially well on a complex problem of natural language disambiguation once they are given enough data.
"These results suggest that we may want to reconsider the tradeoff between spending time and money on algorithm development versus spending it on corpus development".
The idea is that data matters more than algorithms for complex problems (Norvig, 2009).



# Nonrepresentative Training Data

 - In order to generalize well, it is crucial that your trainind data be representative of new cases that you want to generalize to. This is true regardless of whether or not you use instance-based or model-based learning. If you use a nonrepresentative training set, you will train a model that is unlikely to make accurate predictions (overfitting and underfitting are rampant issues here). Sampling Bias happens when you have large samples that are nonrepresentative if sampling method is flawed, and sampling noise happens when there is nonrepresentative data as a result of chance (if sample is too small).
# Poor Quality Data

 - If training data is full of errors, outliers and noise, it will make it harder for ML systems to detect the underlying patterns so the system is less likely to perform well. It is often well worth the effort to spend time cleaning up the training data - most data scientists spend a significant part of their time doing exactly that.
If some instances are clearly outliers, you can discard them or fix the errors manually. If other instances are missing a few features, you must decide whether you want to ignore this attribute altogether, ignore these instances, fill in the missing values or train one model with the feature and one model without it.

# Irrelevant Features
- Garbage In, Garbage Out. Your system will only be capable of learning if the training data contains enough relevant features and not too many irrelevant ones. A critical part of the success of a ML project is coming up with a good set of features to train on (feature engineering).

# Feature engineering generally involves:


1.   Feature Selection (selecting the most useful features to train on among existing features)
2.   Feature Extraction (combining extisting features to produce a more useful one - dimensionality reduction algorithms can help)
3.   Creating new features by gathering data




# Overfitting

* Overgeneralizing because the model performs well on training data but cannot explain anything in the real world beyond the data it was trained on. Generally hppens when the training set is noisy.

* Overfitting happens when the model is too complex relative to the amount and noisiness of the training data, so it starts to learn random patterns in the training data. One can gather more training data, reduce the noise by fixing data errors and removing outliers, or simplifying the model by selecting one with fewer parameters by reducing the number of attributes in the training data or by constraining the model (regularization).







# A Hyperparameter is a parameter of a learning algorithm (not of the model). As such, it is not affected by the learning algorithm itself; it must be set prior to training and remains constant during training. If you set the regularization hyperparameter to a very large value, you will get a flat value (slope is almost zero), so while there is enough bias to ensure overfitting will not happen, your model will also not find a good solution.

# Underfitting

* Model is too simple to learn the underlying structure of the data, because reality is more complex than the model (too much bias means models cannot deal with the variance of the real world)

* You can select a more robust, more powerful model with more parameters
* You can feed better features to the learning algorithm (feature engineering)
* Reduce constraints on the model like reducing regularization hyperparameter.

# Deployment Issues
* Even if you have a large and clean dataset, and you train a model that neither overfits nor underfits the data, you may still run into issues during deployment: for example, the model may be too complex to maintain, or too large to fit into memory, or too slow, or it may not scale properly, it may have security vulnerabilities, it may become outdated if you don't update it often enough, etc.

In short, there is more to an ML project than just data and models. However, the skillset required to handle these operational problems are fairly different from those required for data modeling, which is why companies often have a dedicated MLOps Team to handle this.

# Testing and Validation



*   Splitting Data into a Training Set and Test Set is a good way to monitor how well a model performs before it is deployed to production
*   You train the model using the training set and test it using the test set
* Error Rate on new cases is the generalization error (out of sample error) and by evaluating the model on the test set, you get an estimate of this error. This value tells you how well your model will perform on instances it has never seen before.

If training error is low (model makes few mistakes on the training set) but generalization error is high, it means that your model is overfitting the training data.

* The 80-20% training-test split is common but depends on the size of the dataset: if it contains 10 million instances, then holding out 1% means your test set will contain 100K instances, more than enough to get a good estimate of generalization error.



# How does one choose the value of the regularization hyperparameter?

* You could train 100 different models using 100 different values for ths hyperparameter, but suppose you find the best hyperparameter value that produces a model with the lowest generalization error - say just 5% error. You launch this model into production but unfortunately it does not perform as well as expected and produces 15% errors. What just happened?

* You measured the generalization error multiple times on the test set and you adapted the model and hyperparameters to produce the best model for that particular set - overfitting and does not perform well on new data!


# Holdout Validation



*   A common solution to overfitting when evaluating model performance between different models. You hold out part of the training set to evaluate several candidate models and select the best one. New held-out set is called the validation set (development set). More specifically, you train multiple models with various hyperparameters on the reduced training set (the full training set minus hypervalidation set) and you select the model that performs best on the validation set).
*   After this holdout validation process, you train the best model on the full training set (including the validation set) and this gives you the final model
* Only then do you evaluate this final model on the test set to get an estimate of the generalization error!

This solution usually works quite well. However, if the validation set is too small then the model evaluations will be imprecise: you may end up selecting a suboptimal model by mistake. Conversely if the validation set is too large, then the remaining training set will be much smaller than the full training set. Why is this bad? Well since the final model will be trained on the full training set, it is not ideal to compare candidate models trained on a much smaller training set. It would be like selecting the fastest sprinter to participate in a marathon!

* One way to solve this problem is to perform repeated cross-validation, using many small validation sets. Each model is evaluated once per validation set after it is trained on the rest of the data. By averaging out all the evaluations of a mode, you get a much more accurate measure of its performance. There is a drawback however, the training time is multiplied by the number of validation sets.





# Data Mismatch

In some cases, data used in training is not perfectly representative of the data that will be used in production. Remember that both the validation set and test set must be as representative as possible of the data you expect to use in production, so they should be composed exclusively of representative pictures: you can shuffle them and put half in the validation set and half in the test set (making sure that no duplicates or near-duplicates end up in both sets). After training your model on the web pictures, if you observe that the performance of the model on the validation set is disappointing, you will not know whether this is because your model has overfit the training set, or whether this is just due to the mismatch between the web pictures (what your model trained on) and mobile pictures (what your customers are actually using).

* One solution is to hold out on some of the training pictures (from the web) in yet another set that Andrew Ng calls the train-dev set. After the model is trained on the training set, you evaluate it on the train-dev set. If model performs poorly, then it must have overfit the training set, so you should try to simplify or regularize the model, get more training data, and clean up the training data. But if it performs well on the train-deve set, then you can evaluate the model on the dev set. If it performs poorly then the problem is coming from data mismatch!

* You can engineer features in preprocessing to make them look more like the pictures that will be taken by the mobile app, and then retraining the model. Once you have a model that performs well on both train-dev set and dev set, you can evaluate it one last time on the test set to know how well it is likely to perform in production!

# No Free Lunch Theorem

A model is a simplified representation of data. The simplifications are meant to discard the superfluous details that are unlikely to generalize to new instances. When you select a particular type of model, you are implicitly making assumptions about the data. For example, if you choose a linear model, you are implicitly assuming that the data is fundamentally linear and that the distance between instances and the straight line is just noise (which can be safely ignored).

Wolpert (1996) demonstrated that if you have no assumptions whatsoever about the data, then there is no reason to prefer one model over any other. NFL Theorem states that for some datasets, the best model is a linear mode, while for other datasets it is a neural network. No model is *a priori* guaranteed to work better than all others. So you have to evaluate them all to know for sure. Since this is not possible, in practice you make some reasonable assumptions about the data and evaluate only a few reasonable models.

# Training and running a linear model with Scikit-Learn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
#Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
# Select a linear model
model = LinearRegression()
```

```python
# Train the model
model.fit(X, y)
```

```python
# Make a prediction for Puerto Rico
X_new = [[33_422.8]] # Puerto Rico GDP per capita in 2020
print(model.predict(X_new)) # outputs [[6.01610329]]
```

```python
# Replacing Linear Regression model with K-Nearest Neighbors Regression in the previous code is as easy as replacing:

from sklearn.linear_model import LinearRegression
model = LinearRegression()

#with

from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=3)
```

```python
# Select a 3-Nearest Neighbors regression model
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=3)

# Train the model
model.fit(X, y)

# Make a prediction for Puerto Rico
print(model.predict(X_new)) # outputs [[5.73333333]]
```
